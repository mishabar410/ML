```python
fol1 = "ternovsky"
# fol2 = "ternovsky2"
# fol3 = "ternovsky3"

for i in range(100):
    test_file = tf.io.read_file(fol1 + "/" + str(i) + ".mp3")
    test_audio = tfio.audio.decode_mp3(test_file)

    rows, columns = test_audio.shape
    mas_len = len(test_audio)
    true_audio = tf.reshape(test_audio, [rows * columns])
    np.save("numpy_array/" + fol1 + "_np/" +  str(i) + ".npy", true_audio)
```
